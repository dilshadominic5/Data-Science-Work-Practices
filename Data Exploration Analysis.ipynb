Action performed for null values and outliers for Train Data Set

```python
Q1 = train.quantile(0.25)
Q3 = train.quantile(0.75)
IQR = Q3 - Q1
Features_Out= train[~((train < (Q1 - 3 * IQR)) |(train > (Q3 + 3 * IQR))).any(axis=1)]
Features_Out.shape
```

```python
NumericFeatures.fillna(NumericFeatures.median(), inplace=True)
CategoricalFeatures.fillna(CategoricalFeatures.median(), inplace=True)
DerivedFeatures.fillna(DerivedFeatures.median(), inplace=True)
print(train.isnull().sum())
```
